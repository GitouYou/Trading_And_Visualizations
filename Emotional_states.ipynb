```python
%load_ext nb_black
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation

sns.set()
x = np.array([])
y = np.array([])
y_a = 1000
text = "Ok Let's Buy!"
text1 = "Ok Let's Buy!"
signal = 1
final_y = 0
final_x = 0
fig = plt.figure(figsize=(11, 7))
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
ax.set_title("ETH-USDT", fontsize=20)
ax.set_xlabel("Days After COVID-19 Eradication", fontsize=15)
ax.set_ylabel("Price USDT", fontsize=15)
(line1,) = ax.plot([], [], lw=3)
(line2,) = ax.plot([], [], lw=3, c="r")
(line3,) = ax.plot([], [], lw=3, c="g")
(line4,) = ax.plot([], [], label="toto", ms=10, color="gold", marker="o")


def init():
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    line4.set_data([], [])
    return (
        line1,
        line2,
        line3,
        line4,
    )


def animate(i):
    global x
    global y
    global y_a
    global text
    global text1
    global signal
    global final_x
    global final_y
    x_a = i

    if (y_a > 900) and (signal == 1):
        y_a = y_a + np.random.randint(-18, 12)
        text = "Ok Let's Buy!"

    elif (y_a <= 900) and (y_a > 840) and (signal == 1):
        y_a = y_a + np.random.randint(-6, 4)
        text = "Mmhh... I'm Scared"
    elif (y_a <= 840) and (signal == 1):
        y_a = y_a + np.random.randint(-4, 2)
        text = "I knew this was a bad idea"
        signal = 0
    elif (y_a <= 855) and (signal == 0):
        y_a = y_a + np.random.randint(-6, 10)
    elif (y_a >= 855) and (signal == 0) and (y_a <= 1000):
        y_a = y_a + np.random.randint(-13, 22)
    elif (y_a > 1000) and (signal == 0) and (y_a <= 1150):
        y_a = y_a + np.random.randint(-8, 18)
        text = "I knew everything was going to be Ok"
    elif (y_a > 1250) and (signal != 2):
        text = "T/P, We should sell over here!!"
        final_x = x_a
        final_y = y_a
        y_a = y_a + np.random.randint(-10, 10)
        signal = 2
    elif signal == 2:
        y_a = y_a + np.random.randint(-10, 10)
    else:
        y_a = y_a + np.random.randint(-5, 10)
        signal = 3
        text = "We are Geniuses"
    annotation = plt.annotate("Ok Let's Buy!", xy=(1, 1000), weight="bold", size=25)
    if text1 != text:
        annotation = plt.annotate(text, xy=(x_a, y_a), weight="bold", size=25)
    x = np.append(x, x_a)
    y = np.append(y, y_a)
    line1.set_data(x, y)
    line2.set_data(np.arange(0, 250, 0.1), 800)
    line3.set_data(np.arange(0, 250, 0.1), 1250)
    if signal != 2:
        line4.set_data(x_a, y_a)
    else:
        line4.set_data(final_x, final_y)
        line4.set_markersize(30)
        line1.set_linewidth(2)
        line2.set_color("lightcoral")
    ax.set_xlim(max(0, i - 75), i + 10)
    ax.set_ylim(700, 1350)
    text1 = str(text)
    return line1, annotation, line2, line3, line4


anim = FuncAnimation(fig, animate, init_func=init, frames=230, interval=90, blit=True)


anim.save("test.gif", writer="imagemagick")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
import numpy as np
from celluloid import Camera
sns.set()
numpoints = 10
x=0
y=1000
points = np.array([[x],[y]])
print(points.shape)
colors = np.array(['red'])
sizes = np.array([50])
camera = Camera(plt.figure(figsize=(11,7)))
plt.xlim(-5,65)
plt.title('ETH-USDT',fontsize=20)
plt.xlabel('Days',fontsize=20)
plt.ylabel('USDT',fontsize=20)
for i in range(50):
    x=x+1
    y=y+np.random.uniform(-15,15)
    decission=np.random.uniform(0.3,0.7)
    if decission < 0.4:
        sizes=np.append(sizes,50)
        colors=np.append(colors,'red')
        annotation = plt.annotate("P(s) = "+str(decission)[0:4]+" NO", xy=(x, y), weight="bold", size=20)
        y=y-5
    elif decission < 0.5:
        sizes=np.append(sizes,80)        
        colors=np.append(colors,'orange')
        annotation = plt.annotate("P(s) = "+str(decission)[0:4]+" Don't", xy=(x, y), weight="bold", size=20)
        y=y-2
    elif decission < 0.6:
        sizes=np.append(sizes,110)
        colors=np.append(colors,'yellow')
        annotation = plt.annotate("P(s) = "+str(decission)[0:4]+" Could be", xy=(x, y), weight="bold", size=20)
        y=y+3
    else :
        sizes=np.append(sizes,150)
        colors=np.append(colors,'springgreen')
        annotation = plt.annotate("P(s) = "+str(decission)[0:4]+" Go on!", xy=(x, y), weight="bold", size=20)
        y=y+7
    points=np.append(points,np.array([[x],[y]]),axis=1)
    plt.scatter(*points, c=colors, s=sizes)
    plt.plot(points[0],points[1],color='blue',lw=1.5)
    camera.snap()
anim = camera.animate(blit=True,interval=300)
anim.save('scatter.gif')
```
